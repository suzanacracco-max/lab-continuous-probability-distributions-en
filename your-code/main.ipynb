    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here

#Requirements recap
# - Inputs: bottom, ceiling, count
# - Output: array of uniformly distributed random numbers
# - Remember for SciPy:
# - loc = bottom
# - scale = ceiling - bottom

# 1- Function to generate uniform random numbers

from scipy.stats import uniform

def generate_uniform(bottom, ceiling, count):
    """
    Generate uniformly distributed random numbers.

    Parameters:
    bottom (float): lower boundary
    ceiling (float): upper boundary
    count (int): number of random values to generate

    Returns:
    numpy.ndarray: array of uniform random numbers
    """
    return uniform.rvs(loc=bottom, scale=ceiling - bottom, size=count)


```

```python
# your code here
# 2 Call the function with the given parameters
# First set
uniform_1 = generate_uniform(bottom=10, ceiling=15, count=100)

# Second set
uniform_2 = generate_uniform(bottom=10, ceiling=60, count=1000)
```

```python
# your code here
# 3 - Plot the uniform distributions (histograms)

# Requirements
 # - Histogram
 # - x-axis → value
 # - y-axis → count
 # - bins = 10

import matplotlib.pyplot as plt

# Plot first distribution
plt.hist(uniform_1, bins=10)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Uniform Distribution: 10 to 15 (n=100)")
plt.show()

# Plot second distribution
plt.hist(uniform_2, bins=10)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Uniform Distribution: 10 to 60 (n=1000)")
plt.show()
```

How are the two distributions different?

```python
# your answer below
#Both distributions are uniform, meaning every value within the specified range is equally likely.
#The differences come from range width and sample size.

# 1 - Range of values

   # First distribution:
     # Range: 10 to 15 (width = 5)

   # Second distribution:
     # Range: 10 to 60 (width = 50)
   
 # Effect:
   # The second distribution spreads values across a much wider interval, so each histogram bin covers a larger span of values than in the first distribution.

# 2 - Number of samples (count)
   # First distribution: 100 samples
   # Second distribution: 1,000 samples

 # Effect:
   # With more samples, the second histogram appears smoother and more evenly flat, while the first one may look more irregular due to random variation.
   # This is a classic example of the law of large numbers — as sample size increases, the observed distribution more closely matches the theoretical uniform distribution.

# 3 - Histogram appearance
   # The first histogram:
     # Fewer data points
     # More variability between bin heights
     # Can look “bumpy”

   # The second histogram:
     # More data points
     # Bin counts are more balanced
     # Looks closer to a perfect rectangle

# Both distributions are uniform, but the second distribution spans a wider range and uses a larger sample size, resulting in a smoother and more evenly distributed histogram compared to the first.
```

```python
from scipy.stats import norm 
```

```python
# your code here
# Function to generate normally distributed random numbers

def generate_normal(mean, std, count):
    """
    Generate normally distributed random numbers.

    Parameters:
    mean (float): mean (μ) of the distribution
    std (float): standard deviation (σ)
    count (int): number of random values to generate

    Returns:
    numpy.ndarray: array of normal random numbers
    """
    return norm.rvs(loc=mean, scale=std, size=count)
```

```python
# your code here
# Generate the two datasets

# Normal distribution with std = 1
normal_std_1 = generate_normal(mean=10, std=1, count=1000)

# Normal distribution with std = 50
normal_std_50 = generate_normal(mean=10, std=50, count=1000)
```

```python
# your code here
# Plot the normal distributions

plt.figure(figsize=(12, 4))

# First distribution
plt.subplot(1, 2, 1)
plt.hist(normal_std_1, bins=30)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Normal Distribution (μ=10, σ=1)")

# Second distribution
plt.subplot(1, 2, 2)
plt.hist(normal_std_50, bins=30)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Normal Distribution (μ=10, σ=50)")

plt.tight_layout()
plt.show()
```

```python
# your code here

# Normality test (Shapiro–Wilk)

from scipy.stats import shapiro

stat, p_value = shapiro(data)

print("Test statistic:", stat)
print("p-value:", p_value)
```

Conclusions: 
- Null hypothesis (H₀): data is normally distributed
- If p-value > 0.05 → fail to reject H₀ (looks normal)
- If p-value ≤ 0.05 → reject H₀ (not normal)


With large datasets, the Shapiro test is very sensitive, so even small deviations often lead to rejecting normality.

So it’s completely normal if:

That’s not a contradiction.

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here

# Function to generate exponentially distributed numbers (mean = 10)

def generate_exponential(size, mean=10):
    """
    Generate exponentially distributed random numbers.

    Parameters:
    size (int): number of values to generate
    mean (float): mean of the exponential distribution (default = 10)

    Returns:
    numpy.ndarray: exponentially distributed random numbers
    """
    return np.random.exponential(scale=mean, size=size)
```

```python
# Generate two sequences (size 10 and 100)
exp_10 = generate_exponential(size=10)
exp_100 = generate_exponential(size=100)
```

```python
# Plot the distributions as histograms (bins = 100) side by side

plt.figure(figsize=(12, 4))

# Size = 10
plt.subplot(1, 2, 1)
plt.hist(exp_10, bins=100)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Exponential Distribution (n=10)")

# Size = 100
plt.subplot(1, 2, 2)
plt.hist(exp_100, bins=100)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Exponential Distribution (n=100)")

plt.tight_layout()
plt.show()
```

How are the two distributions different?

1️ - Sample size
  - First distribution: 10 observations
  - Second distribution: 100 observations

Effect:
The distribution with 100 values is much smoother and more clearly shows the exponential shape, while the distribution with 10 values appears noisy and irregular.

2 - Visual appearance
 - n = 10
   - Histogram is sparse
   - Many empty bins
   - Spiky and uneven bars

 - n = 100
   - Bars are more evenly populated
   - Clear right-skewed exponential pattern
   - Long tail is easier to observe

3 Underlying distribution

  - Both datasets are drawn from the same exponential distribution
  - Same mean (10)
  - Same theoretical shape

The difference is not the distribution itself, but how well it is represented by the data.

Key statistical idea
This demonstrates the law of large numbers:
 - As the sample size increases, the empirical distribution more closely approximates the true underlying distribution.


Conclusions: Both datasets are generated from the same exponential distribution with a mean of 10; however, the distribution with 100 samples appears smoother and more clearly exhibits the exponential shape. The distribution with only 10 samples is sparse and irregular due to the small sample size, making the underlying pattern harder to observe.

```python
# Increasing the size, the result is similar to the expected outout
exp_1000 = generate_exponential(size=1000)

plt.hist(exp_1000, bins=100)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Exponential Distribution (n=1000)")
plt.show()
```

The mean changes, so the distribution changes as well.
